2.1 load your training set. Make a train and validation set based on the 80-20 rule.

```python
from pathlib import Path
import numpy as np

train_dir = Path("Reduced MNIST Data/Reduced Trainging data")
test_dir  = Path("Reduced MNIST Data/Reduced Testing data")

def load_paths_labels(base):
    imgs = sorted(base.glob("*/*.jpg"))
    labels = []
    for p in imgs:
        labels.append(int(p.parent.name))
    return np.array(imgs), np.array(labels)

train_paths, train_labels = load_paths_labels(train_dir)
test_paths,  test_labels  = load_paths_labels(test_dir)
```

```python
print("train:", len(train_paths), "images | labels:", np.unique(train_labels))
print("test :", len(test_paths),  "images | labels:", np.unique(test_labels))
```
